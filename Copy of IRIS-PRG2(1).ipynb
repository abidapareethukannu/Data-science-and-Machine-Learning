```python
#import
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
#Load Dataset
iris_bunch=load_iris()
iris=pd.DataFrame(iris_bunch.data,columns=iris_bunch.feature_names)
iris['species']=iris_bunch.target
iris['species']=iris['species'].map(dict(enumerate(iris_bunch.target_names)))
```

```python
#1.Shape and Head
print("Shape of dataset:",iris.shape)
print("\nFirst 5 rows:")
print(iris.head())
```

```python
#2.Basic info
print("\n Data Types and Non_Nulls:")
print(iris.info())
```

```python
#3.Summary Statistics
print("\nSummary Statistics:")
print(iris.describe())
```

```python
#4.Check for missing values
print("\nMissing Values: ")
print(iris.isnull().sum())
```

```python
#5.Class distribution
print("\n Class Distribution:")
print(iris['species'].value_counts())
```

```python
#6.Pairplot for visualizing feature relationships.
sns.pairplot(iris,hue='species',diag_kind='kde')
plt.suptitle("Pairplot or iris features",y=1.02)
plt.show()
```

```python
#7.Corelation matrix
plt.figure(figsize=(8,6))
sns.heatmap(iris.drop('species',axis=1).corr(),annot=True,cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
#8.Boxplots
plt.figure(figsize=(10,6))
sns.boxplot(data=iris.drop('species',axis=1))
plt.title("Boxplot for each feature")
plt.show()
```
