**BREAST CANCER DATASET**

```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

```

```python

# Load Dataset
bc_bunch = load_breast_cancer()
bc = pd.DataFrame(bc_bunch.data, columns=bc_bunch.feature_names)
bc['target'] = bc_bunch.target

```

```python
# 1. Shape and Head
print("Shape of dataset:", bc.shape)
print("\nFirst 5 rows:")
print(bc.head())
```

```python

# 2. Basic info
print("\nData Types and Non_Nulls:")
print(bc.info())
```

```python
# 3. Summary Statistics
print("\nSummary Statistics:")
print(bc.describe())
```

```python
# 4. Check for missing values
print("\nMissing Values:")
print(bc.isnull().sum())
```

```python
# 5. Class distribution
print("\nClass Distribution:")
print(bc['target'].value_counts())
```

```python
# 6. Pairplot (only plotting first few features to keep it readable)
sns.pairplot(bc.iloc[:, :6].assign(target=bc['target']), hue='target', diag_kind='kde')
plt.suptitle("Pairplot of first 6 features", y=1.02)
plt.show()
```

```python
# 7. Correlation matrix
plt.figure(figsize=(12,10))
sns.heatmap(bc.drop('target', axis=1).corr(), annot=False, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
# 8. Boxplots (for first 10 features to avoid clutter)
plt.figure(figsize=(12,6))
sns.boxplot(data=bc.iloc[:, :10])
plt.title("Boxplot for first 10 features")
plt.show()
```
