```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
irisdata=load_iris()
x=irisdata.data
y=irisdata.target
```

```python
print("X Values")
print(x)
print("Y Values")
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
print("X Train Values")
print(x_train)
print("X Test Values")
print(x_test)
```

```python
print("Y Train Values")
print(y_train)
print("Y Test Values")
print(y_test)
```
