DIABETES DATASET

```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
```

```python
# Load the diabetes dataset
diabetes_bunch = load_diabetes()
```

```python
# Convert to DataFrame and add target column
diabetes = pd.DataFrame(diabetes_bunch.data, columns=diabetes_bunch.feature_names)
diabetes['target'] = diabetes_bunch.target
```

```python
# 1. Shape and head
print("Shape of dataset:", diabetes.shape)
print("\nFirst 5 rows:")
print(diabetes.head())
```

```python
# 2. Basic info
print("\nData Types and Non_Nulls:")
print(diabetes.info())
```

```python
# 3. Summary statistics
print("\nSummary Statistics:")
print(diabetes.describe())
```

```python

# 4. Check for missing values
print("\nMissing Values:")
print(diabetes.isnull().sum())
```

```python
# 5. Distribution of target (disease progression)
print("\nTarget (Disease Progression) Distribution:")
print(diabetes['target'].describe())
```

```python
# 6. Pairplot (plotting first 6 features for readability)
sns.pairplot(diabetes.iloc[:, :6].assign(target=diabetes['target']))
plt.suptitle("Pairplot of first 6 features", y=1.02)
plt.show()
```

```python

# 7. Correlation matrix heatmap
plt.figure(figsize=(12,10))
sns.heatmap(diabetes.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()

```

```python

# 8. Boxplots for all features
plt.figure(figsize=(12,6))
sns.boxplot(data=diabetes.iloc[:, :-1])  # Exclude target for boxplot
plt.title("Boxplot for Features")
plt.xticks(rotation=45)
plt.show()
```
