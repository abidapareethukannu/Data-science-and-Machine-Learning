WINE DATASET

```python
# import
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

```python
# Load Dataset
wine_bunch = load_wine()
# wine.data shape: (n_samples, n_features)
wine = pd.DataFrame(wine_bunch.data, columns=wine_bunch.feature_names)
wine['target'] = wine_bunch.target
```

```python
# 1. Shape and Head
print("Shape of dataset:", wine.shape)
print("\nFirst 5 rows:")
print(wine.head())
```

```python
# 2. Basic info
print("\n Data Types and Non_Nulls:")
print(wine.info())
```

```python
# 3. Summary Statistics
print("\nSummary Statistics:")
print(wine.describe())
```

```python
# 4. Check for missing values
print("\nMissing Values:")
print(wine.isnull().sum())
```

```python
# 5. Class distribution
print("\nClass Distribution:")
print(wine['target'].value_counts())
```

```python
# 6. Pairplot (only plotting first few features to keep it readable)
sns.pairplot(wine.iloc[:, :6].assign(target=wine['target']), hue='target', diag_kind='kde')
plt.suptitle("Pairplot of first 6 features", y=1.02)
plt.show()
```

```python
# 7. Correlation matrix
plt.figure(figsize=(12,10))
sns.heatmap(wine.drop('target', axis=1).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
# 8. Boxplots (for first 10 features to avoid clutter)
plt.figure(figsize=(14,6))
sns.boxplot(data=wine.iloc[:, :10])
plt.title("Boxplot for first 10 features")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
