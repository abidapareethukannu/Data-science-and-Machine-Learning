***FOR LOOP***

```python
for x in range(10):
  print(x,end=' ')
limit=int(input('Enter a limit:'))
sum=0
for i in range(1,limit+1):
  if i%2!= 0:
    sum+=i
print("Odd sum="+str(sum))

```

***WHILE LOOP***

```python
number=int(input('Enter number:'))
s=0
while number>0:
  s+=number%10
  number=number//10
print(s)
```

***NESTED LOOPS***

```python
limit=int(input('Enter number:'))
for num in range(2,limit+1):
  is_divisible=False
  k=2
  while k<=num//2:
    if num%k==0:
      is_divisible=True
      break;
    k+=1
  if not is_divisible:
    print(num,end=' ')
```

***CONTAINER-LIST COMPREHENSION***

```python
numbers=list(range(1,8))
print(numbers)
```

```python
square=[]
for i in numbers:
  square.append(pow(i,2))
print(square)
```

***LIST COMPREHENSION***

```python
square=[x**2 for x in numbers]
print(square)
```

***LIST COMPREHENSION WITH A FILTER***

```python
odd_square=[x**2 for x in numbers if x%2!=0]
print(odd_square)
```

```python
A=[4,6,8,9]
AxA=[(a,b)for a in A for b in A if a!=b]
print(AxA)
```
