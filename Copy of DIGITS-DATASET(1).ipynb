```python
# import
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits


```

```python
# Load Dataset
digits_bunch = load_digits()
# digits.data shape: (n_samples, n_features) with 64 features (8x8 pixels)
digits = pd.DataFrame(digits_bunch.data)
digits['target'] = digits_bunch.target

```

```python
# 1. Shape and Head
print("Shape of dataset:", digits.shape)
print("\nFirst 5 rows:")
print(digits.head())
```

```python

# 2. Basic info
print("\n Data Types and Non_Nulls:")
print(digits.info())

```

```python

# 3. Summary Statistics
print("\nSummary Statistics:")
print(digits.describe())

```

```python

# 4. Check for missing values
print("\nMissing Values:")
print(digits.isnull().sum())

```

```python

# 5. Class distribution
print("\nClass Distribution:")
print(digits['target'].value_counts())

```

```python
# 6. Pairplot (only plotting first few features to keep it readable)
sns.pairplot(digits.iloc[:, :6].assign(target=digits['target']), hue='target', diag_kind='kde')
plt.suptitle("Pairplot of first 6 pixel features", y=1.02)
plt.show()

```

```python
# 7. Correlation matrix
plt.figure(figsize=(12,10))
sns.heatmap(digits.drop('target', axis=1).corr(), annot=False, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
```

```python

# 8. Boxplots (for first 10 pixel features to avoid clutter)
plt.figure(figsize=(12,6))
sns.boxplot(data=digits.iloc[:, :10])
plt.title("Boxplot for first 10 pixel features")
plt.show()
```
